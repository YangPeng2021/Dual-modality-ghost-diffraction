```python
sampling_ratio=1

num_patterns=int(np.round(n*m*sampling_ratio))
#num_patterns=1
measurement=np.reshape(measurement,(image_size*image_size,))
measurement=measurement[:num_patterns]
# print(measurement.shape)
# print(pattern_all.shape)
pattern_all=np.zeros((num_patterns,n,m))

k=1
for i in range(num_patterns):   
    #I1=cv2.imread('./pattern_10000.0_butterfly_better/'+str(k).rjust(5,'0')+'.png',0)
    I1=cv2.imread('../transmission/pattern_10000.0_butterfly_final_diffP_shuffle/'+str(k).rjust(5,'0')+'.png',0)
    I1=I1/255
    
    #I2=cv2.imread('./pattern_10000.0_butterfly_better/'+str(k+1).rjust(5,'0')+'.png',0)
    I2=cv2.imread('../transmission/pattern_10000.0_butterfly_final_diffP_shuffle/'+str(k+1).rjust(5,'0')+'.png',0)
    I2=I2/255
    
    pattern_all[i]=I1-I2
    k=k+2
```

```python
from skimage.restoration import denoise_nl_means
def non_local_means(noisy_np_img, sigma, fast_mode=True):
    """ get a numpy noisy image
        returns a denoised numpy image using Non-Local-Means
    """ 
    sigma = sigma / 255.
    h = 0.6 * sigma if fast_mode else 0.8 * sigma
    patch_kw = dict(h=h,                   # Cut-off distance, a higher h results in a smoother image
                    sigma=sigma,           # sigma provided
                    fast_mode=fast_mode,   # If True, a fast version is used. If False, the original version is used.
                    patch_size=5,          # 5x5 patches (Size of patches used for denoising.)
                    patch_distance=6,      # 13x13 search area
                    multichannel=False)
    denoised_img = []
    n_channels = 1
    
    denoise_fast = denoise_nl_means(noisy_np_img[:,:], **patch_kw)
    denoised_img += [denoise_fast]
    return np.array(denoised_img, dtype=np.float32)
```

```python
def ts_to_np(ts):
    
    return ts.detach().cpu().squeeze().numpy()
def np_to_ts(np):
    ts=torch.tensor(np)
    #ts=torch.reshape(ts,(n,m))
    #print("n:",ts)
    #print("v:",ts.unsqueeze(0).size())
    return ts.unsqueeze(0)
```

```python

#pattern_all=(pattern_all-np.mean(pattern_all))/np.std(pattern_all)
pattern_cs_var=np_to_ts(pattern_all).cuda()
pattern_cs_var=torch.reshape(pattern_cs_var,(num_patterns,1,n,m))

measurement=(measurement-np.mean(measurement))/np.std(measurement)
#print(measurement)
measurement_cs_var=np_to_ts(measurement).cuda()
measurement_cs_var=torch.reshape(measurement_cs_var,(1,1,1,num_patterns))

print(measurement_cs_var.shape)
```

```python
def plot_loss_save(k,iters,y_loss,loss):
    #plt.cla()
    iters.append(k)
    loss_np=ts_to_np(loss)
    y_loss.append(loss_np) 

def plot_loss(iters,y_loss):
    plt.plot(iters, y_loss, label='Loss rate', color='g')
    plt.xlabel('Iteration')
    plt.ylabel('MSE')
    plt.title('loss')
    plt.legend()   
    plt.pause(1.0)
```

```python
def tv_loss(x,beta=0.5):
    x=x.cuda()
    dh=torch.pow(x[:,:,:,1:]-x[:,:,:,:-1],2)
    dw=torch.pow(x[:,:,1:,:]-x[:,:,:-1,:],2)
    return torch.sum(dh[:,:,:-1]+dw[:,:,:,:-1])
```

```python
from torch.autograd import Variable
from threading import Thread
import queue
from skimage.restoration import denoise_nl_means, estimate_sigma

num=2000
iters=[]
y_loss=[]
true_loss=[]
lr=0.001
a=0
sigma_f=30
beta=0.5
mu=0.5
update_iter=100
admm_iter=num
x=DGI.copy()
LR_x = 1
f_x,u=x.copy(),np.zeros_like(x)
img_queue = queue.Queue()
#print(img_queue.get())
denoiser_thread = Thread(target=lambda q, f, f_args: q.put(f(*f_args)),
                         args=(img_queue,non_local_means, [x.copy(), sigma_f]))
denoiser_thread.start()      

method='grad'
#avg= np.rint(DGI)
#net = UNet(n_channels=1, n_classes=1, bilinear=False)
net = NAFNet()
criterion = nn.MSELoss().cuda()

if torch.cuda.is_available():
    net.cuda()

optvars = [{'params': net.parameters()}]
optimizier = torch.optim.Adam(optvars, lr=lr,betas=(0.5,0.9),eps=1e-08)

#optimizier = torch.optim.Adam(optvars, lr=lr)
scheduler = torch.optim.lr_scheduler.StepLR(optimizier, step_size=50, gamma=0.9)

if method == 'fixed_point':
    swap_iter = admm_iter + 1
    LR_x = None
elif method == 'grad':
    swap_iter = -1
    LR_x = 1
elif method == 'mixed':
    swap_iter = admm_iter // 2
else:
    assert False, "method can only be 'fixed_point' or 'grad' or 'mixed' !"
        
for k in range(1,1+num): 
    
#     if k == 1:  # the denoiser work in parallel
#         denoiser_thread.join()
#         f_x = img_queue.get()
#         f_x = f_x.reshape((n,m))
        
#         denoiser_thread = Thread(target=lambda q, f, f_args: q.put(f(*f_args)),
#                                  args=(img_queue, non_local_means, [x.copy(), sigma_f_begin]))
#         denoiser_thread.start()
        
    #net_input = net_input + (noise.normal_() * reg_noise_std)
    net_input = net_input.float()
    
    output_net = net(net_input)
    #print(output_net)
    output_net = torch.sigmoid(output_net)
    output = torch.reshape(output_net,(1,n,m))
    
    out_np=ts_to_np(output_net)
    output_net = output_net/torch.max(output_net)
    
    conv=torch.nn.Conv2d(1,num_patterns,(1,1),stride=1)
    conv.weight.data=pattern_cs_var.float()
    conv.bias.data=torch.zeros([num_patterns]).cuda().float()
    
    B_run=conv(output_net)
    B_run=torch.reshape(B_run,(1,1,1,num_patterns))
    
    B_run=(B_run-torch.mean(B_run))/torch.std(B_run)
    #print(B_run)
    
    #print(output.size())
    
    loss_x = criterion(output.float(),np_to_ts(x-u).cuda().float())+a*tv_loss(output_net).float()
    #loss_y = criterion(B_run.float(),measurement_cs_var.float())+a*tv_loss(output_net).float()
    loss_y = criterion(B_run.float(),measurement_cs_var.float())
    total_loss=loss_y+ mu*loss_x
               
    plot_loss_save(k,iters,y_loss,total_loss)
    
    optimizier.zero_grad()

    total_loss.backward()
    scheduler.step()
    optimizier.step()
    
    if k % update_iter == 0:  # the denoiser work in parallel
        denoiser_thread.join()
        f_x = img_queue.get()
        f_x = f_x.reshape((n,m))
        
        denoiser_thread = Thread(target=lambda q, f, f_args: q.put(f(*f_args)),
                                 args=(img_queue, non_local_means, [x.copy(), sigma_f]))
        denoiser_thread.start()
        #print(f_x.shape)
    
    #step 2:
    if k < swap_iter:
        x = 1 / (beta + mu) * (beta * f_x + mu * (out_np + u))  # eq. 11 in the article
    else:
        x = x - LR_x * (beta * (x - f_x) + mu * (x - out_np - u))

    np.clip(x, 0, 1, out=x)
    
    #step 3:
    u=u + out_np - x
    #avg=avg*.99 +out_np *.01
    #print(u.shape)
    #print(x.shape)
    #output = torch.reshape(output_net,(1,n,m))
#     if k % 1000 == 0:
#         net_input=torch.from_numpy(x-u)[None,None, :].cuda()
#         update_iter=30
        
    print ('Iteration %05d, loss %.15f  ' % (k, total_loss.item()), '\r', end='')
    
if denoiser_thread.is_alive():
    denoiser_thread.join()  # joining the thread
```

```python
plot_loss(iters,y_loss)
print("time:",time.process_time()-t0)

result=x-u
#result=out_np
#result[result<0] = 0
print(result.shape)
result=result/np.max(result)
plt.figure(figsize=(4,4))
plt.imshow(result,cmap='gray')
```
